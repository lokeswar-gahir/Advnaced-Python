### Q1. Describe the differences between text and binary files in a single paragraph.
Text files are subset of binary files that are used to store human readable characters as a rich text document or plain text document.These files should be open in "rt" or "wt" mode. Binary files are those typical files that store data in the form of sequence of bytes grouped into eight bits or sometimes sixteen bits. These bits represent custom data and such files can store multiple types of data (images, audio, text, etc) under a single file.These binary files should be opened in "rb" or "wb" mode.
